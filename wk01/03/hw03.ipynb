```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

## Introduction

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [50, 51, 52, 48, 47, 49, 46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x, y, color='green', linewidth=3, linestyle='solid')
```

## Format strings

```python
plt.plot(x, y, 'g+--')
```

```python
plt.plot(x, y, color='green', marker='+', linestyle='dashed', markersize=15)
```

```python
plt.plot(x, y, color='green', alpha=0.3)
```

## Axes labels, Legend, Grid

```python
days = [1, 2, 3, 4, 5, 6, 7]
max_t = [50, 51, 52, 48, 47, 49, 46]
min_t = [43, 42, 40, 44, 33, 35, 37]
avg_t = [45, 48, 48, 46, 40, 42, 41]
```

```python
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.title('Weather')

plt.plot(days, max_t, label='Max')
plt.plot(days, min_t, label='Min')
plt.plot(days, avg_t, label='Avg')

plt.legend(loc='best', shadow=True, fontsize='large')

plt.grid()
```

## Bar Charts

```python
company = ['GOOGL', 'AMZN', 'MSFT', 'FB']
revenue = [90, 136, 89, 27]
profit=[40,2,34,12]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.xticks(ypos, company)
plt.ylabel('revenue (bln)')
plt.title('US Tech Stocks')

plt.bar(xpos - 0.2, revenue, width=0.4, label='Revenue')
plt.bar(xpos + 0.2, profit, width=0.4, label='Profit')

plt.legend()
```

```python
plt.yticks(ypos, company)
plt.title('US Tech Stocks')

plt.barh(xpos - 0.2, revenue, height=0.4, label='Revenue')
plt.barh(xpos + 0.2, profit, height=0.4, label='Profit')
plt.legend()
```

## Histograms

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar, bins=[80, 100, 125, 150], rwidth=0.95, color='green', histtype='step')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('Sugar range')
plt.ylabel('Total no. of patients')
plt.title('Blood sugar analysis')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], rwidth=0.95, color=['green', 'orange'], label=['men', 'women'], orientation='horizontal')
plt.legend()
```

## Pie Chart

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home Rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other Utilities']
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0, 0.2, 0.2, 0, 0], startangle=45)
plt.show()
```

## Save chart to file

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%0.1f%%', shadow=True, explode=[0, 0.2, 0.2, 0, 0], startangle=45)

plt.savefig('piechart.pdf', bbox_inches='tight', pad_inches=2)
```
